```python
import pandas as pd
import numpy as np
# SQL Alchemy
from sqlalchemy import create_engine
from sql_pwd import sql_pwd
# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()
```

```python
connection_string = f'root:{sql_pwd}@localhost/wine_db'
engine = create_engine(f'mysql://{connection_string}')
```

```python
# Query All Records in the the Database

sql = "select * from top_20_wine"
top_20_wine = pd.read_sql_query(sql, con=engine)
```

```python
# Query Single Record in the the Database
top_20_wine_df = pd.DataFrame(top_20_wine, columns=['id', 'wine', 'vintage', 'location','score', 'bottles_in_stock'])
```

```python
top_20_wine_df 
```

```python
more = 'y'

while more=='y':
    input_wine = input("Incoming wine : ")
    input_vintage = input("vintage : ")
    input_bottles = input ("Number of bottles : ")
    nbr_bottles = int(input_bottles)

    for index, row in top_20_wine_df.iterrows():
        if (row["wine"] == input_wine and row["vintage"] == input_vintage):
            top_20_wine_df.loc[index, "bottles_in_stock"]=nbr_bottles
            break;
                
    more = input("Enter another wine delivery ? ")
    
        
```

```python
top_20_wine_df.to_sql(name='top_20_wine', con=engine, if_exists='replace', index=False)
```
