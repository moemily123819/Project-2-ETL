```python
wine_collection_df['vintage']=' '

for index, row in wine_collection_df.iterrows():
    wine_text=[]
    wine_str = row['wine']
    wine_text = wine_str.split(' ')
    if (wine_text[0]>'1900' and wine_text[0]<='2019'):
        wine_collection_df.loc[index, "vintage"] = wine_text[0]
        vintage=wine_text[0]
        wine_2 = wine_str.split(vintage, 1)
        wine_3 = wine_2[1]
        wine_4 = wine_3.lstrip()
        wine_collection_df.loc[index, "wine"] = wine_4
        print(index, wine_text[0], wine_4)
```

```python
#wine_review_url='https://www.wine-searcher.com/find/frogs+leap+sauvignon+blanc/2013/usa/usd'


wine_review_url='https://www.wine-searcher.com/find/'
#wine_string=frogs+leap+sauvignon+blanc/2013/usa/usd'

for index, row in wine_collection_df.iterrows():
    wine_text=[]
    wine_string=''
    wine_str = row['wine']
    wine_text = wine_str.split(' ')
    for y in range(len(wine_text)):
        wine_string= wine_string+wine_text[y]+'+'
    vintage=row["vintage"]    
#   wine_string_1 = f'{wine_string}/{vintage}/usa/usd'
    wine_string_1 = f'{wine_string}/{vintage}'
    
    wine_url = wine_review_url+wine_string_1
    
    print(wine_url)

    executable_path = {'executable_path': 'chromedriver.exe'}
    browser = Browser('chrome', **executable_path, headless=False)


    # HTML object
    time.sleep(60)
    browser.visit(wine_url)
    html = browser.html
    # Parse HTML with Beautiful Soup
    soup = bs(html, 'html.parser')
#    wine_reviews_text = soup.find('meta', name_='description')
#    print(wine_reviews_text)
    desc = soup.find(attrs={"name":"description"})
    content_desc =desc["content"]
    words=[]
    words=content_desc.split(' ')
#    print(len(words))
#    print(words)
    for y in range(len(words)):
        if words[y]=='points.':
            previous_y = y-1
#        print(previous_y, words[previous_y])
            wine_score = words[previous_y]
            wine_collection_df.loc[index, "score"] = wine_score
    
#    print(soup.prettify())
    
    browser.quit()
#    try:
#    # Retrieve all wine shown on the page
#        wine = soup.find_all('td', class_='name')
#          w_list=w.find('h3').text 
#            w_loc = w.find('span', class_="el loc").text
#            print(w_list, w_loc.strip('\n'))
#            print(soup.prettify())
```
